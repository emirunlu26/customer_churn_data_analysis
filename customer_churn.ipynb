# LOAD DATA

```python
import pandas as pd
train_df = pd.read_csv("customer_churn_dataset-training-master.csv")
test_df = pd.read_csv("customer_churn_dataset-testing-master.csv")
```

```python
train_df
```

```python
test_df
```

```python
df = pd.concat([train_df,test_df],ignore_index=True)
df
```

# DATA CLEANING

```python
df.head()
```

## Drop Duplicates

```python
initial_len = len(df.index)
df = df.drop_duplicates()
current_len = len(df.index)
print(f"Number of duplicates removed: {initial_len-current_len}.")
print(f"Number of rows left: {current_len}")
```

## Handle Missing Values

```python
df[(df["CustomerID"].isna()) | (df["Age"].isna()) | (df["Gender"].isna()) | (df["Tenure"].isna())]
```

```python
df[(df["Usage Frequency"].isna()) | (df["Support Calls"].isna()) | (df["Payment Delay"].isna())]
```

```python
df[(df["Subscription Type"].isna()) | (df["Contract Length"].isna()) | (df["Total Spend"].isna())]
```

```python
df[(df["Last Interaction"].isna()) | (df["Churn"].isna())]
```

```python
df = df[df["CustomerID"].notna()]
df.reset_index(drop=True, inplace=True)
```

## Data Type Conversion

```python
df["CustomerID"] = df["CustomerID"].astype("int")
df["Churn"] = df["Churn"].astype("int")
df["Age"] = df["Age"].astype("int")
df["Tenure"] = df["Tenure"].astype("int")
df["Usage Frequency"] = df["Usage Frequency"].astype("int")
df["Support Calls"] = df["Support Calls"].astype("int")
df["Payment Delay"] = df["Payment Delay"].astype("int")
df["Last Interaction"] = df["Last Interaction"].astype("int")
```

```python
df.head()
```

## Drop Irrelevant Columns

```python
df = df.drop("CustomerID", axis=1)
```

```python
df.head()
```

## Handle Outliers

Use Decision Tree for Binary Classification (Customer Churn) -> Model Not Sensitive to Outliers -> Handling Outliers Unnecessary

# ONE-HOT ENCODING

```python
df_encoded = pd.get_dummies(df,dtype=int)
```

```python
df_encoded.head()
```

## Drop One of The Gender Columns to Eliminate Redundancy

```python
df_encoded = df_encoded.drop("Gender_Male", axis=1)
```

```python
df_encoded.head()
```

# TRAIN A DECISION TREE TO PREDICT CUSTOMER CHURN

 ## SPLIT DATASET INTO TRAINING AND TEST SET

```python
input_df = df_encoded.drop(["Churn"],axis=1)
target_df = df_encoded[["Churn"]]
```

```python
input_df.head()
```

```python
target_df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(input_df,target_df,test_size=0.20,random_state=42,shuffle=True)
```

## Hyperparameter Tuning to Select Model With Optimal Hyperparameters

```python
import math

training_sample_num = len(X_train.index)
min_depth = 2
max_depth = math.ceil(math.log2(training_sample_num))

print(f"Min. Depth: {min_depth}")
print(f"Max. Depth: {max_depth}")
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

param_grid = {
    "max_depth": [2,3,5,7,10,12,16,19],
    "criterion": ["gini","entropy"]
}

tree = DecisionTreeClassifier(random_state=42)
grid_search = GridSearchCV(estimator=tree, param_grid=param_grid, cv=5, verbose=True)
```

```python
grid_search.fit(X_train,Y_train)
best_tree = grid_search.best_estimator_
```

```python
print(best_tree)
print(f"Criterion of best tree: {best_tree.criterion}")
```

## Print Features By Their Importance For Churn Prediction

```python
feature_importances = [(feat, round(imp,2)) for feat, imp in zip(X_train.columns, best_tree.feature_importances_)]
feature_importances = sorted(feature_importances, key=lambda x: x[1], reverse=True)
feature_importances
```

## Evaluate Decision Tree (Confusion Matrix, Evaluation Metrics)

```python
Y_train_pred = best_tree.predict(X_train)
Y_test_pred = best_tree.predict(X_test)
```

```python
from sklearn import metrics

accuracy = metrics.accuracy_score(Y_train, Y_train_pred) * 100
print(f"Accuracy of the decision tree on the training set: {accuracy:.2f} %.")
```

```python
from sklearn import metrics

accuracy = metrics.accuracy_score(Y_test, Y_test_pred) * 100
print(f"Accuracy of the decision tree on the test set: {accuracy:.2f} %.")
```

```python
cnf_matrix = metrics.confusion_matrix(Y_test,Y_test_pred)
cnf_matrix
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

labels = [0,1]
sns.heatmap(cnf_matrix, xticklabels=labels, yticklabels=labels, annot=True, cmap="YlGnBu" ,fmt='g')
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.xlabel('Predicted label')
plt.ylabel('Actual label')
```

```python
from sklearn.metrics import classification_report

target_names = ["Negative", "Positive"]
clf_report = classification_report(Y_test, Y_test_pred, target_names=target_names)
print(clf_report)
```

## Subscription Type, Contract Length, Age Group Distribution

```python
# 18-24
# 25-34
# 35-44
# 45-54
# 55-64
# 65+
def age_category(age):
    if age<18:
        return "18-"
    elif age >= 18 and age <=24:
        return "18-24"
    elif age>=25 and age<=34:
        return "25-34"
    elif age>=35 and age<=44:
        return "35-44"
    elif age>=45 and age<=54:
        return "45-54"
    elif age>=55 and age<=64:
        return "55-64"
    else:
        return "65+"
```

```python
df["Age Group"] = df["Age"].apply(age_category)
df.head()
```

## Group Analysis

```python
df.head()
```

```python
subscr_groups = df.groupby("Subscription Type",as_index=False)
spend_by_subscr = subscr_groups.agg(
    total_spend_avg = ("Total Spend","mean")
)
spend_by_subscr["total_spend_avg"] = round(spend_by_subscr["total_spend_avg"],2)
spend_by_subscr = spend_by_subscr.sort_values("total_spend_avg", ascending=False)
spend_by_subscr
```

```python
sns.barplot(x="Subscription Type",y="total_spend_avg",data=spend_by_subscr, palette="pastel",errorbar=None)
plt.title("Average Total Spending by Subscription Type")
plt.xlabel("Subscription Type")
plt.ylabel("Average Total Spending")
plt.show()
```

```python
age_groups = df.groupby("Age Group",as_index=False)
calls_by_age_groups = age_groups.agg(
    support_calls_avg = ("Support Calls","mean"),
    total_spend_avg = ("Total Spend","mean")
)
calls_by_age_groups["support_calls_avg"] = round(calls_by_age_groups["support_calls_avg"],2)
calls_by_age_groups["total_spend_avg"] = round(calls_by_age_groups["total_spend_avg"],2)
calls_by_age_groups.sort_values("support_calls_avg", ascending=False)[["Age Group","support_calls_avg"]]
```

```python
sns.barplot(x="Age Group",y="support_calls_avg",data=calls_by_age_groups, palette="pastel",errorbar=None)
plt.title("Average Support Calls by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Average Support Calls")
plt.show()
```
